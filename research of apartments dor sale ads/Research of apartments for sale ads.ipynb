# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

## Предобработка данных

Заменим пропуски там, где можем предположить их значения. Вероятно, они возникли из-за человеческого фактора - люди забывали указать ту или иную информацию.

```python
#гипотеза: если балкон/ парк/ пруд не указан, значит, его нет
df['balcony'] = df['balcony'].fillna(0) 
df['parks_around3000'] = df['parks_around3000'].fillna(0) 
df['ponds_around3000'] = df['ponds_around3000'].fillna(0) 
df.info() #проверим, что пропуски заполнены
```

```python
#посмотрим на распределение жилой площади в зависимости от количества комнат
print(df.pivot_table(index = 'rooms', values = 'living_area', aggfunc = ['mean', 'median','count']))

#там, где 0 комнат, похоже, что комната 1, близкие значения площадей. Заменим 0 на 1 в количестве комнат
df['rooms'] = df['rooms'].replace(0, 1)

#Пропуски в living_area заменим на медианные значения в соответствии с количеством комнат 
for i in range(df['rooms'].min(),df['rooms'].max()):
    df.loc[(df['rooms'] == i)&(df['living_area'].isna()), 'living_area'] = df[df['rooms'] == i]['living_area'].median()
df.info() #проверим, что пропуски living_area заполнены
```

```python
#Посмотрим на нечисловые значения
print(df['locality_name'].value_counts())# Немало в Ленобласти поселков. Возможно, надо будет объединять в группы 
print(df['first_day_exposition'].unique()) #Да это ж даты! преобразуем их в тип дат

df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'],format='%Y-%m-%dT%H:%M:%S')
```

```python
# Удалим строки, где в столбцах менее 1% данных не заполнено
data = df.copy()
data = data.dropna(subset=['floors_total','locality_name'])
data.info()
```

```python
#В приближении высота потолков не кажется странной: большинство от 2 до 4,5 м. 
#Видимо, дворец из 19 комнат попал. 
#Заменять данные на медианы не будем, выглядят адекватно. Надо будет только отрезать выбросы 
data.plot(y = 'ceiling_height',kind = 'hist',range = (2,5), bins = 50)
```

```python
#Оценим площадь кухни через медианное значение доли кухонь от площадь квартиры.
data['kitchen_share'] = data['kitchen_area']/data['total_area']
print(data['kitchen_share'].median()) #по всем объектам
print(data.pivot_table(index = 'rooms', values = 'kitchen_share', aggfunc='median')) #в разрнезе количества комнат
```

Доля площади кухни все-таки значительно отличается в зависимости от числа комнат. 
Заменим пропуски в площади кухни, учитывая количество комнат и площадь квартиры

```python
for i in range(data['rooms'].min(),data['rooms'].max()):#для объектов с любым количеством комнат
    if len(data.loc[data['rooms'] == i]['rooms'])>0: #если, конечно, такие есть
        #мы заполняем пропуски произведением медианного значения доли кухни и общей площади  квартир с данным количеством комнат 
        data.loc[(data['rooms'] == i)&(data['kitchen_area'].isna()), 'kitchen_area'] = (data.loc[data['rooms'] == i]['kitchen_share'].median())*data['total_area']

data['kitchen_area'].describe()
```

Изменим размерность стоимости квартир: пусть это будут миллионы

```python
data['last_price'] = data['last_price']/1000000
```

Проверим, что этажи логично обозначены

```python
print((data['floor']/data['floors_total']).max())
```

- цену квадратного метра;
- день недели, месяц и год публикации объявления;
