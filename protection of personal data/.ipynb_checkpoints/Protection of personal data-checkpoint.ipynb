## Умножение матриц

Обозначения:

- $X$ — матрица признаков (нулевой столбец состоит из единиц)

- $y$ — вектор целевого признака

- $P$ — матрица, на которую умножаются признаки

- $w$ — вектор весов линейной регрессии (нулевой элемент равен сдвигу)

Предсказания:

$$
a = Xw
$$

Задача обучения:

$$
w = \arg\min_w MSE(Xw, y)
$$

Формула обучения:

$$
w = (X^T X)^{-1} X^T y
$$

**Утверждение:** Качество линейной регрессии при умножении признаков на обратимую матрицу не изменится

**Обоснование:** Если не изменятся предсказания на новом наборе признаков, тогда и качество модели останется без изменений. Проверим, как связаны предсказания, воспользовавшись свойствами произведения матриц:

$$a_1 = XPw_1$$
$$w_1 = ((XP)^T XP)^{-1} (XP)^T y$$
$$w_1 = (P^TX^T XP)^{-1} P^TX^T y$$
$$w_1 = P^{-1}(X^T X)^{-1}(P^T)^{-1} P^TX^T y$$
$$w_1 = P^{-1}(X^T X)^{-1}EX^T y$$
$$w_1 = P^{-1}w$$

$$a_1 = XPP^{-1}w$$
$$a_1 = Xw = a$$

## Алгоритм преобразования

**Алгоритм**

Для кодирования признаков необходимо сформировать случайную матрицу и умножить ее на матрицу признаков

**Обоснование**

В соответстви с доказанной выше леммой, предсказания не изменятся и качество модели не пострадает. При этом мы получим набор случайных векторов, ключом-связкой с исходным набором признаков будет обратная к случайной матрица.

## Проверка алгоритма

```python
#Создадим матрицу признаков

features = df.drop('Страховые выплаты', axis = 1)
target = df['Страховые выплаты']

print(features.shape) #ok, нам нужна матрица 4*4, чтобы умножение прошло успешно
```

```python
#Создадим случайную матрицу и проверим ее на обратимость
random_m = np.random.normal(size=(4, 4))
inv_m = np.linalg.inv(random_m)

display(random_m)
```

```python
#Создадим случайную матрицу и проверим ее на обратимость
def matrix_creation(size_x,size_y):
    random_m = np.random.normal(size=(size_x,size_y))
    try:
        inv_m = np.linalg.inv(random_m)
        return random_m
    except:
        return matrix_creation(size_x,size_y)
    
random_m = matrix_creation(4,4)
inv_m = np.linalg.inv(random_m)

display(random_m)
```

```python
#Закодируем матрицу признаков
features1 = pd.DataFrame (np.dot(features, random_m), columns = features.columns)
```

```python
class LinearRegression:
    def fit(self, train_features, train_target):
        X = np.concatenate((np.ones((train_features.shape[0], 1)), train_features), axis=1)
        y = train_target
        w = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)# < напишите код здесь >
        self.w = w[1:]
        self.w0 = w[0]

    def predict(self, test_features):
        return test_features.dot(self.w) + self.w0
```

```python
model = LinearRegression()

#Сравним результаты модели на двух матрицах признаков
model.fit(features, target)
predictions = model.predict(features)
print('R2 origanal is',round(r2_score(target, predictions),2))

model.fit(features1, target)
predictions = model.predict(features1)
print('R2 coded is',round(r2_score(target, predictions),2))
```

Качество модели до и после преобразования не изменилось.

```python
#Чтобы дешифровать данные, умножим зашифрованную матрицу признаков на обратную (матрицу-ключ) справа
features_back = pd.DataFrame (np.dot(features1, inv_m), columns = features.columns)
```

```python
for i in features_back:
    features_back[i] = round(features_back[i],0)
    print(features_back[i].sum())
```

```python
for i in features:
    print(features[i].sum())
```

Суммы столбцов равны, значит, дешифрование матриц прошло успешно
